```python
# csv로 저장된 제플린 데이터 받아오기
# 최소주문금액 1만원 이하 업소의 계약,  테마코드 데이터 
data = pd.read_csv('mod_oneP_181127.csv',encoding='utf-8') #encoding??
```

```python
# 최소주문금액 1만원 이하 업소 중 1인분 카테고리 켜진업소는 수정x  ->드랍  
# shop_no_to_drop 이라는 업소리스트만든다

shop_no_to_drop = data[data.cd_nm == '1인분']['shop_no']
shop_no_to_drop = shop_no_to_drop.tolist()
shop_no_to_drop[:10]
```

```python
data.reset_index(inplace=True)
```

```python
data.info()
```

```python
to_iter = len(data)
```

```python
len(shop_no_to_drop) 
```

```python
# data의 row만큼 loop돌며 shop_no_to_drop에 포함된 row는 드랍한다
# 최소주문금액이 아직 등록되지 않은 업소 역시 드랍한다

for i in range(to_iter):
    if data.loc[i].shop_no in shop_no_to_drop:
        #print(data.loc[i])
        data.drop(index=data.loc[i][0],inplace=True)
        continue
    if data.loc[i].min_ord_price == 0:
        data.drop(index=data.loc[i][0],inplace=True)
```

```python
data # 처리후의 데이터, shop_no 기준으로 중복처리 필요하다
```

```python
# 일단 파일로 저장해두자
with open("ilinboon_to_on_181127.csv",'w',newline='') as fp:
    wr = csv.writer(fp)
    for i in range(len(data)):
        wr.writerow(data.iloc[i,:])
```

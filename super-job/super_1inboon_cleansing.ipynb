```python
import pandas as pd
import csv
```

```python
# csv로 저장된 제플린 데이터 받아오기
data = pd.read_csv('mod_oneP_181120.csv',encoding='utf-8') #encoding??
```

```python
data.head(10)
```

```python
data.cd_nm = data.cd_nm.astype('category')
```

```python
# data[data.duplicated(keep=False)]
```

```python
# 최소주문금액 1만원 초과 업소 중 1인분 카테고리 켜진업소 shop_no_to_drop
shop_no_to_drop = data[data.cd_nm == '1인분']['shop_no']
shop_no_to_drop = shop_no_to_drop.tolist()
shop_no_to_drop[:10]
```

```python
data.reset_index(inplace=True)
```

```python
data.info()
```

```python
to_iter = len(data)
```

```python
len(shop_no_to_drop) 
```

```python
# data의 row만큼 loop돌며 shop_no_to_drop에 포함된 row는 드랍한다
for i in range(to_iter):
    if data.loc[i].shop_no in shop_no_to_drop:
#         print(data.loc[i])
        data.drop(index=data.loc[i][0],inplace=True)
```

```python
excepted_shop_no = [] # 예외처리된 shop_no

for shop_no in shop_no_unq:
    url = "https://super.smartbaedal.com/admin/shop/shopinfo.asp?shop={}".format(shop_no)
    
    try:
        driver.get(url)

        driver.find_element_by_css_selector("#basic_mod_btn").click() # 수정버튼
        driver.implicitly_wait(0.1)
        driver.find_element_by_css_selector("#basic_form_div > table > tbody > tr:nth-child(5) > td > span > span > a").click() # 테마관리버튼
        driver.implicitly_wait(0.3)
        
        ########### 주의필요
        driver.find_element_by_css_selector("#theme_seq_1").click() # 1인분 테마 클릭( id없으므로 seq 순서 맞는지 확인필요)
        ###########
        
        # 수정하고 저장
        driver.find_element_by_css_selector("body > div:nth-child(12) > div.ui-dialog-buttonpane.ui-widget-content.ui-helper-clearfix > div > button:nth-child(1) > span").click()
        driver.find_element_by_css_selector('#frm > div.content_button > span.wbtn.blue.large.icon.strong > input[type="button"]').click()
        
        # 확인위한 alert 2회
        driver.switch_to.alert.accept()
        driver.switch_to.alert.accept()

        # NoAlertPresentException 발생시(정상적으로 저장안됨) 처리하게 코드짜면 됨
#     except NoAlertPresentException as e1:
#         print("\n\n NoAlertPresentException")
#         print(shop_no, e1)
#         excepted_shop_no.append(shop_no)
#         input("Enter if you checked out")
#         continue
        
    except Exception as e:
        print("\n\n Exception")
        print(shop_no, e)
        excepted_shop_no.append(shop_no)
        input("Enter if you checked out")
        continue
    
```
